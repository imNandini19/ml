```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_index=pd.read_csv("economic_index.csv")
```

```python
df_index.head()
```

```python
# dropping unneccessary features
# here in our data year and month are not required
df_index.drop(columns=["Unnamed: 0","year", "month"], axis=1)
```

```python
df_index
```

```python
# to update the updations in the original data we use inplace=True
df_index.drop(columns=["Unnamed: 0", "year", "month"], axis=1, inplace=True)
df_index
```

```python
# checking if our dataset is having null values
df_index.isnull().sum()
```

```python
import seaborn as sns
sns.pairplot(df_index)
```

```python
df_index.corr()
```

```python
df=df_index
```

```python
# visulazing the datapoints more closely
plt.scatter(df['interest_rate'], df['unemployment_rate'], color="green")
plt.xlabel("Interest Rate")
plt.ylabel("Unemployment Rate")
```

```python
# independent and dependent features
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
import seaborn as sns
```

```python
sns.regplot(x='interest_rate', y='index_price', data=df_index)
```

```python
# sns.regplot(df['interest_rate'], df['index_price'])
sns.regplot(x=df['interest_rate'], y=df['index_price'])
```

```python
sns.regplot(x=df['interest_rate'], y=df['unemployment_rate'])
```

```python
# standard scalar
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
import sklearn
```

```python
from sklearn.model_selection import cross_val_score
validation_score=cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
```

```python
validation_score
```

```python
np.mean(validation_score)
```

```python
# prediction
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
# performance metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse=mean_squared_error(y_test, y_pred)
mae=mean_absolute_error(y_test, y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test, y_pred)
print(score)
```

```python
plt.scatter(y_test, y_pred)
```

```python
residuals=y_pred-y_test
print(residuals)
```

```python
# plotting the residuals
sns.displot(residuals, kind="kde")
```

```python
# scatter plot with respect to prediction and residuals
plt.scatter(y_pred, residuals)
```
