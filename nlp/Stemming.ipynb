**Porter stemmer**

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.stem import PorterStemmer, RegexpStemmer, SnowballStemmer
```

```python
ps=PorterStemmer()
rgs=RegexpStemmer("ing|s$|e$|ation$|able")
sbs=SnowballStemmer('english')
```

```python
sentence="The organizations are studying various connections and creating sustainable solutions."
words=nltk.word_tokenize(sentence)
words
```

```python
print("Porter stemmer")
for word in words:
  print(word, "----->", ps.stem(word))
print("\n\nRegexp stemmer")
for word in words:
  print(word, "----->", rgs.stem(word))
print("\n\nSnow ball stem")
for word in words:
  print(word, "----->", sbs.stem(word))
```
