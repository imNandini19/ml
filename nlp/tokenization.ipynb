```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus="""Hello guys!! welcome to my youtube channel.
This is your favourite youtuber nandini. make sure to like subscribe to my channel. i love you."""
```

```python
print(corpus)
```

```python
documents=sent_tokenize(corpus)
```

other than new line it is also giving that after any explanation mark also a new

```python
type(documents)
```

```python
for sentence in documents:
  print(sentence)
```

```python
# word tokenization
# paragraph / sentence ===> words
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
